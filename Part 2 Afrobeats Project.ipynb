```python
#Scraping festival lineups
#Starting with Glastonbury 2022

from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
url = "https://www.glastonburyfestivals.co.uk/line-up/line-up-2022/?artist"
html = requests.get(url).text
```

```python
soup = BeautifulSoup(html)
```

```python
soup.select(".title")
```

```python
artistsGB22 = soup.select(".title")
GB22 = []
for title in artistsGB22:
    Glastonbury22_df = {
        "artist": title.text
    }
    GB22.append(Glastonbury22_df)
```

```python
GB22
```

```python
Glastonbury22_df = pd.DataFrame(GB22)
```

```python
Glastonbury22_df
```

```python
Glastonbury22_df = Glastonbury22_df.drop_duplicates()
```

```python
Glastonbury22_df
```

```python
# Create a set of artists to match
artists_set = {"BURNA BOY", "CUPPY", "DENISE CHAILA",
               "K.O.G", "KOJEY RADICAL", "LES AMAZONES D'AFRIQUE",
               "MOONCHILD SANELLY", "PA SALIEU", "PONGO",
               "SAMPA THE GREAT", "TEMS"}

# Apply the transformation using the apply method and a lambda function
Glastonbury22_df["artist"] = Glastonbury22_df["artist"].apply(lambda x: 1 if x in artists_set else 0)
Glastonbury22_df.to_csv('Glastonbury22.csv', index=False)
```

```python
#Scraping the Afrobeats Artist Sites from Last FM to check them with the festival lineups
from bs4 import BeautifulSoup
url2 = "https://www.last.fm/tag/afrobeats/artists?page=1"

html2 = requests.get(url2).text

soup = BeautifulSoup(html2)
```

```python
soup.select(".big-artist-list-title")
```

```python
#All artists on page one

for artist in soup.select(".big-artist-list-title"):
    print(artist.text)
```

```python
#Now I want to loop through all the pages

base_url = "https://www.last.fm/tag/afrobeats/artists?page="

artist_fm = []

for page in range(1, 26):
    url = base_url + str(page)
    req = requests.get(url)
    print(f"Scraping url{url}")

    if req.status_code != 200:
        print(f"Fehler beim Abrufen der Seite {page}: Statuscode {req.status_code}")
        continue
    
    soup = BeautifulSoup(req.text, 'html.parser')
    artists = soup.find_all(class_ = "big-artist-list-title")
    print(f"Seite {page}: {len(artists)} Künstler gefunden")  # Debugging-Ausgabe
    
    for i in range(20):
        artist_fm.append(artists[i].text)
        # if page>1:
        #     print(f"{(i)+page*20}" + artists[i].text)
        # else:
        print(f"{i}" + artists[i].text)
```

```python
artist_fm
```

```python
#Putting everything in upper case like in the Glastonbury dataset
artists_fm_df["artists_fm_name"] = artists_fm_df["artists_fm_name"].str.upper()
```
